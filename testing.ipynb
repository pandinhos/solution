Импортируем библиотеки NumPy, Pandas.

```python
import pandas as pd
import numpy as np
```

Загрузим файл dataset_Facebook.csv и не забудем изменить сепаратор по умолчанию. Выведем для проверки несколько последних значений.

```python
data = pd.read_csv('dataset_Facebook.csv', sep = ';')
data.tail(5)
```

Посчитаем среднее значение по всему датасету.

data.mean()

Вычислим минимальное значение по всем столбцам для всех типов контента.

```python
data.min()
```

А теперь максимальное значение.

```python
data.max()
```

Найдем второй квартиль для каждого столбца, то есть медиану. 

```python
data.median()
```

Так как несколько значений встречаются одинаковое количество раз, а также при этом максимальное количество раз, то в некоторых колонках будет сразу несколько модальных значений. В случае отсутствия моды (если значение не встречается чаще одного раза) нам вернется значение NaN.

```python
data.mode()
```

Узнаем сколько различных значений принимает столбец 'Type'.

```python
data['Type'].unique()
```

Извлечем подвыборки из датасета для каждого типа контента.

```python
photo = data.loc[data['Type'] == 'Photo']
status = data.loc[data['Type'] == 'Status']
link = data.loc[data['Type'] == 'Link']
video = data.loc[data['Type'] == 'Video']
```

Исключим столбец 'Type', так как методы mean, median для объекта типа DataFrame не учитывают нечисловые переменные.

```python
columns = list(data.columns.values)
columns.remove('Type')
```

Таблица средних значений по столбцам для всего датасета и различных типов контента.

```python
mean = pd.DataFrame({
    'Column': columns,
    'Dataset': data.mean().values, 
    'Photo': photo.mean().values,
    'Status': status.mean().values,
    'Link': link.mean().values,
    'Video': video.mean().values
                    })
mean
```

Таблица минимальных значений по столбцам для всего датасета и различных типов контента.

```python
minim = pd.DataFrame({
    'Column': data.columns.values,
    'Dataset': data.min().values, 
    'Photo': photo.min().values,
    'Status': status.min().values,
    'Link': link.min().values,
    'Video': video.min().values
                    })
minim
```

Таблица максимальных значений по столбцам для всего датасета и различных типов контента.

```python
maxim = pd.DataFrame({
    'Column': data.columns.values,
    'Dataset': data.max().values, 
    'Photo': photo.max().values,
    'Status': status.max().values,
    'Link': link.max().values,
    'Video': video.max().values
                    })
maxim
```

Таблица медиан по столбцам для всего датасета и различных типов контента.

```python
median = pd.DataFrame({
    'Column': columns,
    'Dataset': data.median().values, 
    'Photo': photo.median().values,
    'Status': status.median().values,
    'Link': link.median().values,
    'Video': video.median().values
                    })
median
```
